```python

# Core libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# ML libraries
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                              classification_report, confusion_matrix, roc_auc_score)

# Advanced boosting libraries (install if not already available)
# !pip install xgboost lightgbm catboost
import xgboost as xgb
import lightgbm as lgb
import catboost as cb

# Deep learning
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Reproducibility
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
def iqr_outlier_summary(df, cols):
    summary = {}
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        n_outliers = ((df[col] < lower) | (df[col] > upper)).sum()
        summary[col] = {'lower_bound': lower, 'upper_bound': upper,
                         'n_outliers': n_outliers, 'pct_outliers': 100 * n_outliers / len(df)}
    return pd.DataFrame(summary).T

iqr_outlier_summary(train, sensors)
```

```python
train['target'] = pd.to_numeric(train['target'], errors='coerce').astype(int)   # ← naya add

train['Date'] = pd.to_datetime(train['Date'])
test['Date'] = pd.to_datetime(test['Date'])

print("Train date range:", train['Date'].min(), "to", train['Date'].max())
print("Test date range:", test['Date'].min(), "to", test['Date'].max())
print("Unique timestamps in train:", train['Date'].nunique())

daily_anomaly_rate = train.groupby(train['Date'].dt.date)['target'].mean()
plt.figure(figsize=(14, 4))
daily_anomaly_rate.plot()
plt.title('Anomaly Rate Over Time')
plt.show()
```

```python
for col in sensors:
    if train[col].isnull().sum() > 0:
        median_val = train[col].median()
        train[col] = train[col].fillna(median_val)
        test[col] = test[col].fillna(median_val)

print("Missing values after imputation (train):", train[sensors].isnull().sum().sum())
print("Missing values after imputation (test):", test[sensors].isnull().sum().sum())
```

```python
n_dupes = train.duplicated().sum()
print(f"Duplicate rows found: {n_dupes}")
if n_dupes > 0:
    train = train.drop_duplicates().reset_index(drop=True)
    print("New train shape after dropping duplicates:", train.shape)
```

```python
def cap_outliers(df, cols, bounds):
    df = df.copy()
    for col in cols:
        lower, upper = bounds[col]
        df[col] = df[col].clip(lower=lower, upper=upper)
    return df

bounds = {}
for col in sensors:
    Q1 = train[col].quantile(0.25)
    Q3 = train[col].quantile(0.75)
    IQR = Q3 - Q1
    bounds[col] = (Q1 - 3 * IQR, Q3 + 3 * IQR)  # wider bound (3*IQR) to avoid clipping real anomalies too aggressively

train_capped = cap_outliers(train, sensors, bounds)
test_capped = cap_outliers(test, sensors, bounds)
```

```python
def engineer_features(df, sensors, is_train=True):
    df = df.sort_values('Date').reset_index(drop=True)

    # Time-based features
    df['hour'] = df['Date'].dt.hour
    df['day'] = df['Date'].dt.day
    df['dayofweek'] = df['Date'].dt.dayofweek
    df['month'] = df['Date'].dt.month

    # Rolling statistics (window=5) per sensor — captures local trend/volatility
    for col in sensors:
        df[f'{col}_roll_mean5'] = df[col].rolling(window=5, min_periods=1).mean()
        df[f'{col}_roll_std5'] = df[col].rolling(window=5, min_periods=1).std().fillna(0)

    # Lag features — previous reading and delta from previous reading
    for col in sensors:
        df[f'{col}_lag1'] = df[col].shift(1).fillna(df[col].iloc[0])
        df[f'{col}_diff1'] = df[col] - df[f'{col}_lag1']

    # Cross-sensor interaction features
    df['X1_X2_ratio'] = df['X1'] / (df['X2'] + 1e-6)
    df['X4_X5_diff'] = df['X4'] - df['X5']
    df['sensor_sum'] = df[sensors].sum(axis=1)
    df['sensor_std'] = df[sensors].std(axis=1)

    return df

train_fe = engineer_features(train_capped, sensors)
test_fe = engineer_features(test_capped, sensors)

print("Train features shape:", train_fe.shape)
train_fe.head()
```

```python
feature_cols = [c for c in train_fe.columns if c not in ['Date', 'target', 'ID']]
print(f"Total features used: {len(feature_cols)}")
print(feature_cols)
```

```python
X = train_fe[feature_cols]
y = train_fe['target']

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

print("Train set:", X_train.shape, " Val set:", X_val.shape)
print("Train target balance:\n", y_train.value_counts(normalize=True))
print("Val target balance:\n", y_val.value_counts(normalize=True))

# Scale features for distance/gradient-based models (LogReg, KNN, SVM, NN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
results = []

def evaluate_model(name, y_true, y_pred, y_proba=None):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_proba) if y_proba is not None else np.nan

    results.append({'Model': name, 'Accuracy': acc, 'Precision': prec,
                     'Recall': rec, 'F1': f1, 'ROC_AUC': auc})

    print(f"--- {name} ---")
    print(classification_report(y_true, y_pred, target_names=['Normal', 'Anomaly']))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Anomaly'], yticklabels=['Normal', 'Anomaly'])
    plt.title(f'{name} — Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
# XGBoost
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()  # handles imbalance

xgb_model = xgb.XGBClassifier(
    n_estimators=300, max_depth=6, learning_rate=0.05,
    scale_pos_weight=scale_pos_weight, eval_metric='logloss',
    random_state=SEED, n_jobs=-1
)
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_val)
y_proba = xgb_model.predict_proba(X_val)[:, 1]
evaluate_model('XGBoost', y_val, y_pred, y_proba)
```

```python
# LightGBM
lgb_model = lgb.LGBMClassifier(
    n_estimators=300, max_depth=8, learning_rate=0.05,
    class_weight='balanced', random_state=SEED, n_jobs=-1
)
lgb_model.fit(X_train, y_train)
y_pred = lgb_model.predict(X_val)
y_proba = lgb_model.predict_proba(X_val)[:, 1]
evaluate_model('LightGBM', y_val, y_pred, y_proba)
```

```python
# CatBoost
cb_model = cb.CatBoostClassifier(
    iterations=300, depth=6, learning_rate=0.05,
    auto_class_weights='Balanced', random_state=SEED, verbose=0
)
cb_model.fit(X_train, y_train)
y_pred = cb_model.predict(X_val)
y_proba = cb_model.predict_proba(X_val)[:, 1]
evaluate_model('CatBoost', y_val, y_pred, y_proba)
```

```python
param_grid = {
    'max_depth': [4, 6, 8],
    'n_estimators': [200, 300],
    'learning_rate': [0.03, 0.05, 0.1]
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=SEED)

grid_search = GridSearchCV(
    estimator=xgb.XGBClassifier(scale_pos_weight=scale_pos_weight,
                                 eval_metric='logloss', random_state=SEED, n_jobs=-1),
    param_grid=param_grid, scoring='f1', cv=cv, n_jobs=-1, verbose=1
)
grid_search.fit(X_train, y_train)

print("Best params:", grid_search.best_params_)
print("Best CV F1 score:", grid_search.best_score_)

best_xgb = grid_search.best_estimator_
y_pred = best_xgb.predict(X_val)
y_proba = best_xgb.predict_proba(X_val)[:, 1]
evaluate_model('XGBoost (Tuned)', y_val, y_pred, y_proba)
```

```python
nn_model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# class weights for imbalance
class_weight = {0: 1.0, 1: scale_pos_weight}

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = nn_model.fit(
    X_train_scaled, y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=50, batch_size=512,
    class_weight=class_weight,
    callbacks=[early_stop],
    verbose=1
)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].plot(history.history['loss'], label='train_loss')
axes[0].plot(history.history['val_loss'], label='val_loss')
axes[0].set_title('Loss'); axes[0].legend()

axes[1].plot(history.history['accuracy'], label='train_acc')
axes[1].plot(history.history['val_accuracy'], label='val_acc')
axes[1].set_title('Accuracy'); axes[1].legend()
plt.show()
```

```python
y_proba_nn = nn_model.predict(X_val_scaled).flatten()
y_pred_nn = (y_proba_nn >= 0.5).astype(int)
evaluate_model('Neural Network', y_val, y_pred_nn, y_proba_nn)
```

```python
# Residual analysis — look at predicted probability vs actual class to spot systematic errors
residual_df = pd.DataFrame({
    'actual': y_bt_test.values,
    'predicted_proba': y_bt_proba
})
residual_df['residual'] = residual_df['actual'] - residual_df['predicted_proba']

plt.figure(figsize=(10, 4))
plt.scatter(range(len(residual_df)), residual_df['residual'], alpha=0.3, s=5)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals (Actual − Predicted Probability) over Backtest Period')
plt.ylabel('Residual')
plt.show()

print("Mean residual:", residual_df['residual'].mean())
print("Residual std:", residual_df['residual'].std())
```
